```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pytesseract
```

```python
# Cell 2: Load and display the text image

# Replace 'your_image.png' with your image filename
image_path = "/content/IMage.jpeg"

# Read image using OpenCV
img = cv2.imread(image_path)

# Convert BGR (OpenCV default) to RGB for correct plotting
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image
plt.figure(figsize=(8,6))
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Input Text Image")
plt.show()
```

```python
# Cell 3: Preprocessing and OCR

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding (binary image)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display preprocessed image
plt.figure(figsize=(8,6))
plt.imshow(thresh, cmap='gray')
plt.axis("off")
plt.title("Preprocessed Image")
plt.show()

# Perform OCR using pytesseract
extracted_text = pytesseract.image_to_string(thresh)

print("Extracted Text from Image:")
print("---------------------------------")
print(extracted_text)
```
